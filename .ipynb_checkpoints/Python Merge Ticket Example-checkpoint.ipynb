## Import Pandas Library and Data

```python
# Import
import pandas as pd
```

```python
# import data into df
east_df = pd.read_csv("/Users/morgan/Documents/Courses/Python/MyModules/east.csv")
west_df = pd.read_csv("/Users/morgan/Documents/Courses/Python/MyModules/west.csv")
north_df = pd.read_csv("/Users/morgan/Documents/Courses/Python/MyModules/north.csv")
south_df = pd.read_csv("/Users/morgan/Documents/Courses/Python/MyModules/south.csv")
staff_df = pd.read_csv("/Users/morgan/Documents/Courses/Python/MyModules/staff.csv")
tickets_df = pd.read_csv("/Users/morgan/Documents/Courses/Python/MyModules/tickets.csv")
```

## Concatinate The Stations

```python
stations_df = pd.concat([east_df, west_df, north_df, south_df])
```

```python
stations_df
```

```python
stations_df.shape
```

## Merge Stations with Tickets

```python
result_df = pd.merge(tickets_df,
                 stations_df[['Barcode', 'Location','Staff']],
                  on='Barcode',
                  how='left')
```

```python
# Group the data by location
location_group = result_df.groupby(['Location'])
```
