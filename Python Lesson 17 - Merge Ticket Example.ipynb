## Concatinate The Stations

```python
# Use Pandas concat() method
stations_df = pd.concat([east_df, west_df, north_df, south_df])
```

## Merge  Tickets with Stations
We now wish to merge the data from the tickets data frame with the Stations Data Frame, to discover what stations the the ticket holders went to.

```python
# Output a new daraframe with the merged data merge().
result_df = pd.merge(tickets_df,
                 stations_df[['Barcode', 'Location','Staff',"Subject"]],
                  on='Barcode',
                  how='left')
```

```python
# Group the data by location
location_group = result_df.groupby(['Location'])
```

```python
# Group the data by location and Profession
profession_group = result_df.groupby(['Location','Profession'])
```

```python
# Filter to find out what doctors went to the East station
profession_group.get_group(('East','Doctor'))
```

```python
# Get a count of each profession that presented at each station.
profession_group[['Location', 'Profession']].count()
```
